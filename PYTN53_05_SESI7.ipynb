# BASIC VISUALIZATION

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_canada = pd.read_excel("data/Canada.xlsx", sheet_name="Canada by Citizenship",
                          skiprows=range(20), skipfooter=2)
df_canada
```

```python
df_canada = pd.read_excel("data/Canada.xlsx", sheet_name="Canada by Citizenship",
                          skiprows=range(20), skipfooter=2)
df_canada
```

```python
df_canada.drop(columns=["Type", "Coverage", "AREA", "REG", "DEV"], inplace=True)
df_canada.rename(columns={"OdName": "Country", "AreaName": "Area", "RegName": "Region"}, inplace=True)
df_canada
```

```python
df_canada["Total"] = df_canada.sum(axis=1)
df_canada
```

```python
df_canada.Country.nunique()
```

```python
df_canada.set_index("Country", inplace=True)
df_canada
```

```python
print(df_canada.loc["Japan", range(1980, 2000)])
print(df_canada.loc["Indonesia", range(1980, 2000)])
```

```python
years = list(range(1980, 2014))
print(years)
```

```python
df_canada.loc[df_canada.Region == "South-Eastern Asia", years]
```

```python
df_top5 = df_canada.sort_values("Total", ascending=False)
df_top5 = df_top5.head(5)
df_top5 = df_top5[years].T

df_top5.plot(kind="area", alpha=.8, figsize=(18, 7))
plt.xlabel("Years")
plt.ylabel("Num of Immigrants")
plt.title("Top 5 Countries")
plt.xticks(years[::2])
plt.show()
```

```python
df_bottom5 = df_canada.sort_values("Total", ascending=False)
df_bottom5 = df_bottom5.tail(5)
df_bottom5 = df_bottom5[years].T

df_bottom5.plot(kind="area", alpha=.45, figsize=(10, 8))
plt.xlabel("Years")
plt.ylabel("Num of Immigrants")
plt.title("Top 5 Countries Contribute The Least Immigrants")
plt.xticks(years[::2], rotation=90)
plt.show()
```

```python
df_canada[2013].plot(kind="hist", figsize=(8, 6))
plt.title("Immigrants Distribution of 195 Countries in 2013")
plt.xlabel("Num of Immigrants")
plt.ylabel("Num of Countires")
plt.show()
```

```python
freq, bins = np.histogram(df_canada[2013], bins=20)
df_canada[2013].plot(kind="hist", figsize=(12, 6), bins=20)
plt.title("Immigrants Distribution of 195 Countries in 2013")
plt.xlabel("Num of Immigrants")
plt.ylabel("Num of Countires")
plt.xticks(bins, rotation=90)
plt.show()
```

```python
plt.hist(df_canada[2013])
```

```python
freq, bin_range = np.histogram(df_canada.loc[["Denmark", "Norway", "Sweden"], years], bins=15)
df_canada.loc[["Denmark", "Norway", "Sweden"], years].T.plot(
    kind="hist",
    alpha=.5,
    bins=15,
    xticks=bin_range,
    figsize=(10, 6),
)
plt.xticks(rotation=45)
plt.title("Immigrants Distribution from Denmark, Norway, Sweden")
plt.xlabel("Num of Immigrants")
plt.ylabel("Num of Years")
plt.show()
```

```python
iceland = df_canada.loc["Iceland", years]
iceland.plot(kind="bar", figsize=(10, 5))
plt.title("Immigrants from Iceland")
plt.xlabel("Year")
plt.ylabel("Num of Immigrants")

plt.annotate(
    "",
    xy=(32, 70),
    xytext=(26.5, 20),
    xycoords="data",
    arrowprops={"arrowstyle": "->", "connectionstyle": "arc3", "color": "blue", "lw": 2}
)

plt.annotate(
    "2008-2011 Financial Crisis",
    xy=(26, 25),
    rotation=62,
    va="bottom",
    ha="left",
)

plt.show()
```

```python
df_top15 = df_canada.sort_values("Total", ascending=False)["Total"].head(15)
df_top15.plot(kind="barh", figsize=(8, 10))
plt.title("Top 15 Countries Contributing to Immigration to Canada")
plt.xlabel("Number of Immigrants")
plt.show()
```

```python
df_continents = df_canada.groupby("Area").sum()
df_continents["Total"].plot(
    kind="pie",
    figsize=(15, 6),
    autopct="%1.1f%%",
    labels=None,
    pctdistance=1.2,
    explode=[0.1, 0, 0, 0, 0.1, 0.1]
)
plt.axis("equal")
plt.title("Immigration by Area (1980 - 2013)")
plt.legend(labels=df_continents.index, loc="upper left")
plt.show()
```

```python
japan = df_canada.loc[["Japan"], years].T
japan.plot(kind="box", figsize=(7, 5))
plt.title("Summary Statistics of Japanese Immigrants")
plt.ylabel("Num of Immigrants")
plt.show()
```

```python
df_tot = df_canada[years].sum()
df_tot = df_tot.reset_index()
df_tot.columns = ["year", "total"]
df_tot
```

```python
df_tot.plot(kind="scatter", x="year", y="total")
plt.title("Total Immigrants to Canada")
plt.xlabel("Year")
plt.ylabel("Num of Immigrants")
plt.show()
```

```python
fig = plt.figure(figsize=(15, 6))
ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)

china_india.T.plot(kind="box", vert=False, ax=ax0)
ax0.set_title("Box plots of Immigrants from China and India", fontdict={"fontsize": 12})
ax0.set_xlabel("Num of Immigrant")
ax0.set_ylabel("Countries")

china_india.T.plot(kind="line", ax=ax1)
ax1.set_title("Immigrants Growth of China and India to Canada", fontdict={"fontsize": 12})
ax1.set_xlabel("Years")
ax1.set_ylabel("Num of Immigrants")
plt.show()
```
